```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 1000
duration = 1
t = np.linspace(0, duration, int(fs * duration), endpoint=False)
```

```python
t[:10]
```

```python
f0 = 5
x = np.sin(2 * np.pi * f0 * t)
```

```python
x[:10]
```

```python
plt.figure(figsize=(10, 4))
plt.plot(t, x)
plt.title("5 Hz Sine Wave Sampled at 1000 Hz")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
theta = 2 * np.pi * f0 * t

x_circle = np.cos(theta)
y_circle = np.sin(theta)

plt.figure(figsize=(6, 6))
plt.plot(x_circle, y_circle, label="Phasor path (unit circle)")
plt.scatter(x_circle, y_circle, c=theta, cmap='twilight', s=100)
plt.xlabel("cos(0) = x")
plt.ylabel("sin(0) = y")
plt.gca().set_aspect('equal')
plt.grid(True)
plt.title("Rotating Phasor on Unit Circle")
plt.colorbar(label="Angle 0 (radians)")
plt.legend()
plt.show()
```

```python
fs = 100
f0 = 1
duration = 1
t = np.linspace(0, duration, int(fs * duration), endpoint=False)
theta = 2 * np.pi * f0 * t

x_circle = np.cos(theta)
y_circle = np.sin(theta)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Unit circle (phasor path)
axs[0].plot(x_circle, y_circle, label="Phasor path")
axs[0].scatter(x_circle, y_circle, c=theta, cmap='twilight', s=20)
axs[0].set_title("Unit Circle (Phasor Rotation)")
axs[0].set_xlabel("cos(θ)")
axs[0].set_ylabel("sin(θ)")
axs[0].set_aspect('equal')
axs[0].grid(True)

# Sine wave (y projection)
axs[1].plot(t, y_circle, label="sin(θ) = y", color='tab:blue')
axs[1].scatter(t, y_circle, c=theta, cmap='twilight', s=20)
axs[1].set_title("Sine Wave = Y-Projection of Phasor")
axs[1].set_xlabel("Time (s)")
axs[1].set_ylabel("Amplitude")
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(t, y_circle, label="sin(θ) — y-projection")
plt.plot(t, x_circle, label="cos(θ) — x-projection", linestyle='--')
plt.axhline(0, color='black', lw=0.5)
plt.title("Sine and Cosine from Phasor Rotation")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```
